## Organização dos dados

Para agrupar os dados coletados usamos dicionários. A estrutura pode ser entendida assim:
```js
{
    seq: {
        full: {
            Tamanho de entrada(2^4...2^13): Pair<Média, Intervalo de confiança>
        },
        seahorseValley: ... ,
        elephantValley: ... ,
        tripleSpiralValley: ...
    },
    seqio: ...Igual a seq,
    pth: {
        full: {
            Tamanho de entrada(2^4...2^13): [
                2^0 threads: Pair<Média, Intervalo de confiança>,
                2^1 threads: Pair<Média, Intervalo de confiança>,
                ... ,
                2^5 threads: Pair<Média, Intervalo de confiança>
            ]
        },
        seahorseValley: ... ,
        elephantValley: ... ,
        tripleSpiralValley: ...
    },
    omp: ...Igual a pthread
}
```
Onde *seq* representa o tempo total e *seqio* representa o tempo sem as operações de I/O e gerenciamento de memória da versão sequencial.

````julia
data = Dict();
data["seq"] = Dict("full"=>Dict(), "seahorseValley"=>Dict(), "elephantValley"=>Dict(), "tripleSpiralValley"=>Dict());
data["seqio"] = Dict("full"=>Dict(), "seahorseValley"=>Dict(), "elephantValley"=>Dict(), "tripleSpiralValley"=>Dict());
data["pth"] = Dict("full"=>Dict(), "seahorseValley"=>Dict(), "elephantValley"=>Dict(), "tripleSpiralValley"=>Dict());
data["omp"] = Dict("full"=>Dict(), "seahorseValley"=>Dict(), "elephantValley"=>Dict(), "tripleSpiralValley"=>Dict());

keys = ["full", "seahorseValley", "elephantValley", "tripleSpiralValley"];
for i in 4:13
     for t in keys
         pair = 1 => 0;
         data["seq"][t][2^i] = pair;
         data["seqio"][t][2^i] = pair;
         data["pth"][t][2^i] = Pair{Float64, Float64}[];
         data["omp"][t][2^i] = Pair{Float64, Float64}[];
     end
end
````

## Intervalo de confiança

Como o intervalo de confiança é simétrico em torno da média, só armazenamos os limiares. Para 95% de nível de confiança e 9 graus de liberdade, obtemos pela tabela da t-student o valor 2.2622 para z.


````julia
function confidence_interval(vector, mean)
    z = 2.2622;
    n = length(vector);
    sigma = 0;
    for i in 1:n
        sigma += (vector[i] - mean)^2;
    end
    sigma /= n-1;
    sigma = sqrt(sigma);
    return z * sigma/sqrt(n);
end;
````

````julia
# Processamento dos dados
size = nrow(csv)

for i in 11:10:size+1
    tipo = csv.tipo[i-1];
    figura = csv.figura[i-1];
    tamanho = csv.tamanho[i-1];
    threads = csv.threads[i-1];
    if (tipo == "seq")
        vector = vcat(csv.tempo[i-10:i-1], csv.io[i-10:i-1]);
        sum = reduce(+, vector);
        pair = sum/10 => confidence_interval(vector, sum/10);
        data["seqio"][figura][tamanho] = pair;

        sum  = reduce(+, csv.tempo[i-10:i-1]);
        pair = sum/10 => confidence_interval(csv.tempo[i-10:i-1], sum/10);
        data[tipo][figura][tamanho] = pair;
    else
        sum  = reduce(+, csv.tempo[i-10:i-1]);
        pair = sum/10 => confidence_interval(csv.tempo[i-10:i-1], sum/10);
        push!(data[tipo][figura][tamanho], pair);
    end
end
````

````julia
# Definição do eixo x
x = [];
for i in 4:13
    append!(x, 2^i);
end
````

````julia
# Definição de fontes para o plot
axis_font = Plots.font("Helvetica", 7);
text_font = Plots.font("Helvetica", 10);
````

````julia
function fetch_data(tipo, id)
    plot_seq = [];
    ci_seq = [];
    plot_seqio = [];
    ci_seqio = [];
    plot_pth = [];
    ci_pth = [];
    plot_omp = [];
    ci_omp = [];
    for i in 4:13
        append!(plot_seq, data["seq"][tipo][2^i].first);
        append!(plot_seqio, data["seqio"][tipo][2^i].first);
        append!(plot_pth, data["pth"][tipo][2^i][id].first);
        append!(plot_omp, data["omp"][tipo][2^i][id].first);

        append!(ci_seq, data["seq"][tipo][2^i].second);
        append!(ci_seqio, data["seqio"][tipo][2^i].second);
        append!(ci_pth, data["pth"][tipo][2^i][id].second);
        append!(ci_omp, data["omp"][tipo][2^i][id].second);
    end
    
    return plot_seq, ci_seq, plot_seqio, ci_seqio, plot_pth, ci_pth, plot_omp, ci_omp;
end;
````

````julia
function plot_by_number_of_threads(threads)
    id = convert(Int64, log2(threads) + 1);
    for tipo in keys
        plot_seq, ci_seq, plot_seqio, ci_seqio, plot_pth, ci_pth, plot_omp, ci_omp = fetch_data(tipo, id);
        
        title = "\n" * string(threads) * " Thread(s) - " * tipo;
        
        seq = plot(x, plot_seq, label= "Seq", lw = 1, legend=:bottomright,
            color="#7a5306", fillcolor="#f4b22e",
            xaxis = :log, yaxis=(:log, [0.1, :auto]),
            seriestype = :scatter,
            ribbon = ci_seq)
        
        seqio = plot(x, plot_seqio, label= "Seq sem I/O", lw = 1, legend=:bottomright,
            color="#810ee0", fillcolor="#d382ab",
            xaxis = :log, yaxis=(:log, [0.1, :auto]),
            seriestype = :scatter,
            ribbon = ci_seqio)
        
         pth = plot(x,  plot_pth, label= "Pth", lw = 1, legend=:bottomright,
            color="#000396", fillcolor="#07ffc0",
            xaxis = :log, yaxis=(:log, [0.1, :auto]),
            seriestype = :scatter,
            ribbon = ci_pth)
        
         omp = plot(x, plot_omp, label= "OMP", lw = 1, legend=:bottomright,
            color="#f60000", fillcolor="#29a7e1",
            xaxis = :log, yaxis=(:log, [0.1, :auto]),
            seriestype = :scatter,
            ribbon = ci_omp)
        
        final_plot = plot(seq, seqio, pth, omp, size=(750, 410), layout=4,
            guidefont=axis_font, xtickfont=axis_font, ytickfont=axis_font,
            titlefont=text_font, legendfont = axis_font,
            bottom_margin = 7mm,
            title=title, xlabel="Tamanho da entrada", ylabel="Tempo(ms)")
        
        display(final_plot)
    end
end;
````
