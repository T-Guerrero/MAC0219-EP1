```julia
using CSV
using DataFrames
using Plots

csv = CSV.read("../fake.csv", DataFrame)
x = 1:5;
seq = csv.seq;
pth = csv.pth;
omp = csv.omp;
plot(x, seq, title="Abuble", label= "seq", lw = 3, size=(800, 600), legend=:bottomright)
plot!(x, pth,label="pthreads", lw = 3)
plot!(x, omp,label="OMP", lw = 3)
xlabel!("x")
ylabel!("y")
```
